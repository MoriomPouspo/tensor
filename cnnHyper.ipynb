```python
import tensorflow as tf
from keras import datasets, layers, models
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
#from keras.wrappers.scikit_learn import KerasClassifier
#from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from scikeras.wrappers import KerasClassifier


# Load CIFAR-10 dataset
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
# Specify the number of train images you want to take
num_train_images = 5000
num_test_images = 5000

# Take the specified number of train images and labels
train_images = train_images[:num_train_images]
train_labels = train_labels[:num_train_images]

# Take the specified number of train images and labels
test_images = test_images[:num_test_images]
test_labels = test_labels[:num_test_images]

# Normalize pixel values
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
# Model function with expanded hyperparameter options
def create_model(optimizer='adam', filters=32, dense_units=64, learning_rate=0.001, batch_size=32):
  model = models.Sequential()
  model.add(layers.Conv2D(filters, (3, 3), activation='relu', input_shape=(32, 32, 3)))
  # ... (rest of the model architecture)
  model.compile(optimizer=optimizer,
                loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                metrics=['accuracy'])
  return model
```

```python

# Create KerasClassifier
keras_model = KerasClassifier(model=create_model, verbose=1, learning_rate=0.005, filters=64, dense_units=64)

```

```python
# Define a wider hyperparameter space
param_grid = {
    'optimizer': ['adam', 'sgd', 'rmsprop'],
    'filters': [32, 64, 128],
    'dense_units': [64, 128, 256],
    'learning_rate': [0.001, 0.005, 0.01],
    'batch_size': [32, 64, 128]
}

# Employ RandomizedSearchCV for more efficient exploration
grid = RandomizedSearchCV(estimator=keras_model, param_distributions=param_grid, cv=3, n_iter=10)

```

```python
# Perform hyperparameter tuning
grid_result = grid.fit(train_images, train_labels)

```

```python
# Summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
# ... (rest of the result analysis)
```
